```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, f1_score
import matplotlib.pyplot as plt
```

```python
def sigmoid(X):
    return 1 / (1 + np.exp(-X))

def heaviside(X):
    return np.heaviside(X - 0.5, 0)

def f1(preds):
    logits, y_true = preds
    y_pred = heaviside(sigmoid(logits))
    return {'f1':f1_score(y_true, y_pred, average='macro')}
```

```python
ex = 11
train_pred = pd.read_csv(f'experiments/ex{ex}/results/train_pred.csv')
train_true = pd.read_csv(f'experiments/ex{ex}/results/train_true.csv')
val_pred = pd.read_csv(f'experiments/ex{ex}/results/val_pred.csv')
val_true = pd.read_csv(f'experiments/ex{ex}/results/val_true.csv')
zhihu_pred = pd.read_csv(f'experiments/ex{ex}/results/zhihu_pred.csv')
zhihu_true = pd.read_csv(f'experiments/ex{ex}/results/zhihu_true.csv')
```

```python
res11 = classification_report(val_true, heaviside(sigmoid(val_pred)), output_dict=True)
lp11 = train_true.mean()
res11 = [(key, res11[key]['f1-score']) for key in res11.keys()]
```
